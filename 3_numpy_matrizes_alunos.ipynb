## 📖 Bloco Teórico + Analogias (40 min)

### O que são Arrays N-dimensionais?

Na Aula 2 você trabalhou com arrays **1D** — uma sequência linear de números, como uma única coluna de uma planilha. Hoje subimos um degrau: arrays **2D** (matrizes), que têm linhas e colunas — como a planilha inteira.

O NumPy suporta arrays com qualquer número de dimensões (por isso "N-dimensional"):

| Dimensões | Nome usual | Analogia |
|-----------|-----------|---------|
| 1D | Array / Vetor | Uma coluna de planilha |
| 2D | Matriz | A planilha inteira (linhas × colunas) |
| 3D | Cubo de dados | Várias planilhas empilhadas |
| 4D+ | Tensor | Cubos ao longo do tempo (vídeo, por exemplo) |

> 🏢 **Analogia do prédio:** Um array 1D é um corredor (só tem comprimento). Um array 2D é um andar do prédio (tem comprimento e largura). Um array 3D é o prédio inteiro (comprimento, largura e altura). O NumPy consegue trabalhar com qualquer número de "dimensões" do prédio.

Em Data Science, você vai trabalhar principalmente com 1D e 2D. Mas saber que o conceito existe para mais dimensões é importante — quando você trabalhar com imagens ou redes neurais, vai encontrar arrays 3D e 4D.

### Criação de Matrizes

Há cinco formas principais de criar matrizes:

#### `np.array()` com lista de listas

A forma mais direta: cada lista interna vira uma linha da matriz. Todas as listas internas devem ter o mesmo tamanho.

> 📊 **Analogia da planilha:** É como digitar uma tabela no Excel linha por linha. Cada lista interna é uma linha digitada.




### Passo 1: Criando matrizes com `np.array()` 2D

````python
import numpy as np

vendas = np.array([
    [100, 150, 200, 250],    # Produto A : semanas 1, 2, 3, 4
    [120, 180, 240, 230],    # Produto B
    [130, 195, 260, 280]     # Produto C
])

print(vendas)
print(type(vendas))
````

---

### Atributos das Matrizes

Os mesmos quatro atributos da Aula 2 funcionam para matrizes, mas agora com informações bidimensionais:

| Atributo | Array 1D | Matriz 2D |
|----------|----------|-----------|
| `.shape` | `(5,)` | `(3, 4)` → 3 linhas, 4 colunas |
| `.ndim` | `1` | `2` |
| `.dtype` | `int64` | `float64` |
| `.size` | `5` | `12` → 3 × 4 |

> 📐 **Regra de leitura do shape:** Em `(3, 4)`, o primeiro número é sempre o número de **linhas** e o segundo é o número de **colunas**. Pense: "3 linhas com 4 elementos cada".

---



### Passo 2: Inspecionando atributos da matriz

````python
print("shape: ", vendas.shape)
print("ndim: ", vendas.ndim)
print("dtype: ", vendas.dtype)
print("size: ", vendas.size)
````

### Passo 3: Criando matrizes especiais

#### `np.zeros()` | `np.ones()` | `np.full()` com tupla de dimensões

Agora, em vez de passar um número (como na Aula 2), você passa uma **tupla** com `(linhas, colunas)` para criar uma matriz inteira preenchida com zeros ou uns.



````python
mat_zeros = np.zeros((4,3))
print("Matriz de zeros: \n", mat_zeros)

mat_ones = np.ones((4,3))
print("Matriz de um: \n", mat_ones)

mat_full = np.full((4,3), 7)
print("Matriz de sete: \n", mat_full)
````

#### `np.eye()` — Matriz Identidade

Cria uma matriz quadrada com 1s na diagonal principal e 0s em todo o resto. Muito usada em álgebra linear e algoritmos de Machine Learning.

> 🪞 **Analogia do espelho:** A matriz identidade é o "neutro" da multiplicação matricial — multiplicar qualquer matriz pela identidade retorna a própria matriz, assim como multiplicar qualquer número por 1 retorna o próprio número.



````python
mat_ident = np.eye(4)
print("Matriz identidade: \n", mat_ident)
````

#### `np.random.rand()` — Matriz de valores aleatórios

Cria uma matriz preenchida com números aleatórios entre 0 e 1. Indispensável para inicializar pesos de modelos, gerar dados de teste e simulações.

> 🎲 **Analogia do dado:** É como lançar um dado de infinitas faces para cada célula da tabela e anotar o resultado. Cada vez que você executa, os valores mudam.

````python
#np.random.seed(42)  # Para resultados reprodutíveis
mat_aleatoria = np.random.rand(4,3)
print("Matriz aleatória: \n", mat_aleatoria.round(2))
````

````python
mat_alet = np.random.uniform(0, 10, (4,3))
print("Matriz aleatória entre 0 e 10: \n", mat_alet.round(2))
````

### `reshape()` e `flatten()`

**`reshape()`** reorganiza os dados de um array em novas dimensões, **sem alterar os valores** nem a quantidade de elementos. É como reorganizar os móveis de um apartamento: os móveis são os mesmos, só a disposição muda.

> 🧩 **Analogia do quebra-cabeça:** Você tem 12 peças. Pode organizá-las em 3 fileiras de 4, ou em 4 fileiras de 3, ou em 2 fileiras de 6 — as peças são sempre as mesmas, só o formato da mesa muda. A única regra: o total de peças deve ser o mesmo.

**`flatten()`** faz o caminho inverso: transforma qualquer matriz de volta em um array 1D. É como empilhar todas as linhas de uma planilha em uma única coluna longa.

> 📜 **Analogia do rolo de papiro:** Você pega uma planilha retangular (2D) e a "enrola" em um papiro linear (1D).

---



### Passo 4: `reshape()`  — reorganizando dimensões

````python
seq = np.arange(1,13)
print("Sequência de 1 a 12: ", seq)
print("shape: ", seq.shape)

matriz_3x4 = seq.reshape(3,4)
print("Matriz 3x4: \n", matriz_3x4)

matriz_4x3 = seq.reshape(4,3)
print("Matriz 4x3: \n", matriz_4x3)

matriz_2x = seq.reshape(2, -1)
print("Matriz 2x: \n", matriz_2x)

matriz_2x = seq.reshape(-1, 3)
print("Matriz 2x: \n", matriz_2x)



````

### Indexação e Fatiamento em Matrizes

Em arrays 1D, o índice era um único número. Em matrizes 2D, são **dois números separados por vírgula**: `matriz[linha, coluna]`.

- **Um elemento:** `matriz[2, 3]` → linha 2, coluna 3
- **Uma linha inteira:** `matriz[1, :]` → linha 1, todas as colunas (o `:` significa "tudo")
- **Uma coluna inteira:** `matriz[:, 2]` → todas as linhas, coluna 2
- **Uma submatriz:** `matriz[0:2, 1:3]` → linhas 0-1, colunas 1-2

> 🗺️ **Analogia do mapa do tesouro:** Para encontrar um tesouro no mapa, você precisa de dois valores: latitude e longitude. `matriz[linha, coluna]` funciona igual — dois "endereços" para chegar a um elemento específico.

### Passo 6: Indexação e fatiamento em matrizes

````python
#" Matriz de notas: 4 alunos e 3 provas"
notas = np.array([
    [8.5, 7.0, 9.0],   # Aluno 0 : provas 1, 2, 3
    [6.0, 8.0, 7.5],   # Aluno 1
    [9.0, 9.5, 10.0],  # Aluno 2
    [5.5, 6.0, 6.5]    # Aluno 3
])

# Acessando um único elemento da matriz (Nota do Aluno 1 na Prova 2)
val_unico = notas[1,1]
print("Nota do Aluno 1 na Prova 2: ", val_unico)

# Acessando uma linha inteira
print("notas do Aluno 2: ", notas[2, :])

# Acessando uma coluna inteira
print("Notas da Prova 3: ", notas[:,2])

#submatriz: Notas dos Alunos 0 e 1 nas Provas 2 e 3
submatriz = notas[0:2, 1:3]
print("Submatriz (Alunos 0 e 1, Provas 2 e 3): \n", submatriz)

# Usando índice negativo para acessar a última prova do último aluno
ultima_nota = notas[-1, -1]
print("Última nota (Aluno 3, Prova 3): ", ultima_nota)

print("Ultima linha: ", notas[-1, :])
````

### Operações Matriciais

As operações entre matrizes de mesmo tamanho funcionam **element-wise** (elemento a elemento), assim como nos arrays 1D:

- `A + B` → soma elemento por elemento
- `A - B` → subtração elemento por elemento
- `A * B` → multiplicação elemento por elemento (**NÃO é o produto matricial!**)

O **produto matricial** (multiplicação "de verdade" entre matrizes, como na matemática) é feito com `np.dot(A, B)` ou com o operador `@`.

> ⚽ **Analogia do futebol:** `A * B` é como comparar dois times jogador por jogador, posição por posição (zagueiro vs zagueiro, atacante vs atacante). `np.dot(A, B)` é o jogo de verdade — cada jogador de um time interage com todos os do outro time.

**Regra do produto matricial:** Para multiplicar A por B com `np.dot()`, o número de **colunas de A** deve ser igual ao número de **linhas de B**.

---



### Passo 7: Operações matriciais element-wise e broadcasting

````python
import numpy as np

# Faturamento de 3 lojas em 2 semestres
faturamento = np.array([
    [150, 180],         #Loja A : 1º e 2º semestre
    [200, 220],         #Loja B
    [130, 160]          #Loja C
])

#Custo por semestre para cada loja
custo = np.array([
    [80, 90],
    [110, 115],
    [70, 85]
])

# Resultado do Lucro por semestre para cada loja
lucro = faturamento - custo
print ("Lucro por semestre para cada loja: \n", lucro)

# Aplicar uma meta de crescimento de 10% no faturamento.
meta = faturamento *1.10
print("Meta de crescimento de 10% no faturamento: \n", meta)

# Estatísticas por eixo
# axis=0: Operação ao longo das colunas (por semestre)
# axis=1: Operação ao longo das linhas (por loja)
print("\n Total por semestre (axis=0): ", np.sum(faturamento, axis=0))
print("\n Total por loja (axis-1):",  np.sum(faturamento, axis=1))
print("\n Média por loja (axis-1):",  np.mean(faturamento, axis=1))


````

### Passo 8: Produto matricial com `np.dot()`


````python
# Cenário: calcular receita total das lojas
# Matriz de quantidades vendidas: 2 x 3 (2 lojas, 3 produto)

quantidade = np.array([
    [100, 50, 200],     # Loja A : Produto 1, 2, 3
    [80, 70, 150]       # Loja B

])

# Criando arrya de preços (vetor)
precos = np.array([25, 40, 15])

# Produto  matricial: (2, 3) x (3,)
# Cada loja: soma de (qtd_prod1 * preco_prod1) + (qtd_prod2 * preco_prod2) + (qtd_prod3 * preco_prod3
receita = np.dot(quantidade, precos)
print("Receita total por loja: ", receita)


# Multiplicação de matrizes vs produto matricial
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

produto_elementwise = A * B
print("Produto elementwise: \n", produto_elementwise)

produto_matricuial = np.dot(A, B)
print("Produto matricial (dot): \n", produto_matricuial)


# Primeira linha de A * Primeira coluna de B: (1*5) + (2*7) = 5 + 14 = 19
# Primeira linha de A * Segunda coluna de B: (1*6) + (2*8) = 6 + 16 = 22
# Segunda linha de A * Primeira coluna de B: (3*5) + (4*7) = 15 + 28 = 43
# Segunda linha de A * Segunda coluna de B: (3*6) + (4*8) = 18 + 32 = 50






````

### Transposição: `.T`

Transpor uma matriz significa trocar linhas por colunas. O que era a linha 0 vira a coluna 0, o que era a linha 1 vira a coluna 1, e assim por diante. Uma matriz `(3, 4)` transposta vira `(4, 3)`.

> 🔄 **Analogia da planilha virada:** Imagine uma planilha com meses nas colunas e lojas nas linhas. Transpor é como "virar a planilha de lado": agora os meses ficam nas linhas e as lojas nas colunas. Os dados são os mesmos — só a perspectiva mudou.

Em Data Science, transpor é essencial para alinhar dimensões antes de multiplicações matriciais e para reformatar dados para entrada em modelos.


### Passo 9: Transposição com `.T`

````python
# Matriz de alunos po disciplina: 3 alunos e 4 disciplinas

notas_alunos = np.array([
    [8.5, 7.0, 9.0, 6.5],   # Aluno 0 : disciplinas 1, 2, 3, 4
    [6.0, 8.0, 7.5, 9.0],   # Aluno 1
    [9.0, 9.5, 10.0, 8.0]   # Aluno 2
])

print("Matriz Original de Notas dos Alunos: \n", notas_alunos)
print("Shape original da matrix: ", notas_alunos.shape)

transp = notas_alunos.T
print("Matriz transposta de Notas dos Alunos: \n", transp)
print("Shape original da matrix: ", transp.shape)
````

### ⚠️ Erros Frequentes de Iniciantes

**Erro 1: Passar uma lista simples (não lista de listas) para criar uma matriz**

```python
# ❌ ERRADO: passar uma lista plana quando quer uma matriz
errado = np.array([1, 2, 3, 4, 5, 6])
print(errado.shape)   # (6,) → é 1D, não 2D!

# ✅ CORRETO: usar lista de listas para criar matriz 2D
correto = np.array([[1, 2, 3],
                    [4, 5, 6]])
print(correto.shape)  # (2, 3) → isso sim é uma matriz
```

```
(6,)
(2, 3)
```

---

**Erro 2: Listas internas com tamanhos diferentes**

```python
# ❌ ERRADO: listas internas de tamanhos diferentes
errado = np.array([[1, 2, 3],
                   [4, 5]])    # Linha com 2 elementos quando deveria ter 3!
# Resultado: VisibleDeprecationWarning e dtype=object (não é matriz numérica!)
print(errado.dtype)   # object — problema!

# ✅ CORRETO: todas as linhas com o mesmo número de elementos
correto = np.array([[1, 2, 3],
                    [4, 5, 0]])  # Preencha com 0 se necessário
print(correto.dtype)  # int64
```

```
object
int64
```

---

**Erro 3: `reshape()` com número de elementos incompatível**

```python
arr = np.arange(12)   # 12 elementos

# ❌ ERRADO: tentar fazer reshape para um shape incompatível
# arr.reshape(3, 5)   # 3 × 5 = 15 ≠ 12 → ValueError!

# ✅ CORRETO: o produto das dimensões deve ser igual ao total de elementos
arr_ok = arr.reshape(3, 4)   # 3 × 4 = 12 ✅
arr_ok2 = arr.reshape(2, 6)  # 2 × 6 = 12 ✅
print(arr_ok.shape, arr_ok2.shape)
```

```
(3, 4) (2, 6)
```

---

**Erro 4: Confundir `A * B` com `np.dot(A, B)`**

```python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[2, 0],
              [1, 3]])

# ❌ CONFUSÃO: usar * quando quer o produto matricial matemático
print("A * B (element-wise):\n",  A * B)

# ✅ CORRETO: usar np.dot() para o produto matricial "de verdade"
print("np.dot(A, B) (produto matricial):\n", np.dot(A, B))
```

```
A * B (element-wise):
 [[ 2  0]
 [ 3 12]]
np.dot(A, B) (produto matricial):
 [[ 4  6]
 [10 12]]
```

---

**Erro 5: Tentar acessar elemento com índice único em vez de dois índices**

```python
matriz = np.array([[10, 20, 30],
                   [40, 50, 60]])

# ❌ CONFUSÃO: achar que matriz[1] retorna um elemento
print(matriz[1])      # Retorna a LINHA 1 inteira: [40 50 60], não um elemento!

# ✅ CORRETO: usar dois índices para chegar a um único elemento
print(matriz[1, 2])   # Linha 1, coluna 2 → 60
```

```
[40 50 60]
60
```
