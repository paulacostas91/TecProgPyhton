## 🔥 Aquecimento — Problema Gerador 

Imagine que você trabalha há três semanas em um notebook de análise de dados. Você fez uma limpeza nos dados, calculou médias, criou gráficos incríveis — o trabalho está quase pronto. Então, na madrugada antes da entrega, você decide "melhorar" uma parte do código. Apaga algumas linhas, reescreve outras… e tudo para de funcionar. O notebook dá erro. Você tenta desfazer, mas já salvou por cima. **O trabalho de três semanas foi pelo ralo.**

Soa familiar? Isso acontece com analistas de dados o tempo todo.

Agora imagine uma segunda situação: você e mais dois colegas estão trabalhando no mesmo projeto. Cada um edita o notebook no computador de casa e manda por e-mail: `analise_v1.ipynb`, `analise_v2_joao.ipynb`, `analise_FINAL.ipynb`, `analise_FINAL_mesmo.ipynb`. Qual é o mais atualizado? Quem mudou o quê? Como juntar tudo sem perder nada?

**Essas duas situações têm a mesma solução: o GIT.**

O GIT é um sistema de controle de versão — um "histórico eterno" do seu código. Pense nele como o **Ctrl+Z perfeito**, que nunca expira, funciona mesmo depois de fechar o computador, guarda quem fez cada alteração e permite que várias pessoas trabalhem no mesmo projeto sem pisar no serviço do outro.

Em Data Science, onde projetos evoluem constantemente e equipes são a regra, dominar o GIT não é opcional — é o primeiro passo antes de escrever uma linha de Python.

## 📖 Bloco Teórico + Analogias 

### O que é GIT e por que versionar código?

**GIT** é um sistema de controle de versão distribuído, criado em 2005 pelo mesmo programador que criou o Linux. Ele registra cada alteração feita nos seus arquivos ao longo do tempo, permitindo voltar a qualquer versão anterior, comparar mudanças e trabalhar em equipe sem conflitos.

> 📚 **Analogia do caderno de rascunho:** Imagine que você está escrevendo um trabalho de escola num caderno. O GIT é como tirar uma fotografia do caderno cada vez que você termina uma página importante. Se você errar nas próximas páginas, pode voltar e copiar exatamente do que estava na foto. Sem o GIT, você apaga e perde para sempre.

**Por que é essencial em Data Science?**

- Projetos de dados mudam muito: novos dados chegam, modelos são revisados, análises são refeitas.
- Times de dados colaboram: cientistas, engenheiros e analistas trabalham nos mesmos arquivos.
- Revisões são necessárias: professores, gestores e clientes pedem alterações. Com GIT, nada se perde.
- Portfólio no GitHub: recrutadores da área olham seu repositório antes de te chamar para entrevista.

---

### Instalação e Configuração Inicial

O GIT precisa ser instalado uma vez no seu computador. Após a instalação, você precisa dizer ao GIT quem você é — nome e e-mail — para que cada alteração fique registrada com sua autoria.

> 🪪 **Analogia do crachá:** É como colocar seu crachá antes de entrar numa empresa. Todo registro de ponto (commit) vai ter seu nome. Se você não usar crachá, ninguém sabe quem trabalhou.

A configuração é feita com dois comandos simples no terminal, e você só precisa fazer isso uma vez por computador.

---

### O Workflow do GIT: as Quatro Zonas

O GIT organiza o seu trabalho em quatro "lugares" distintos. Entender esses lugares é a chave para nunca mais se perder nos comandos:

| Zona | O que é | Analogia |
|------|---------|---------|
| **Working Tree** | Seus arquivos no computador, onde você trabalha | A bancada do açougue — onde você corta e prepara |
| **Staging Area** | Uma "lista de compras" do que vai ser salvo | A bandeja onde você separa o que vai levar ao caixa |
| **Repositório Local** | O histórico de versões salvo no seu computador | O livro-caixa da loja, guardado no cofre |
| **Repositório Remoto** | O histórico salvo na nuvem (GitHub) | A filial online da loja, acessível de qualquer lugar |

> 🛒 **Analogia do supermercado:** Você pega produtos nas prateleiras (working tree), coloca no carrinho (staging area), passa no caixa e paga (commit → repositório local), e depois manda entregar na sua casa (push → repositório remoto). Você não manda entregar o supermercado inteiro — só o que está no carrinho.

O caminho sempre segue essa ordem:
```
Working Tree → (git add) → Staging Area → (git commit) → Repositório Local → (git push) → Repositório Remoto
```

---

### Comandos Essenciais

Cada comando do GIT corresponde a uma ação nesse fluxo:

| Comando | O que faz |
|---------|-----------|
| `git init` | Cria um novo repositório GIT na pasta atual |
| `git status` | Mostra quais arquivos foram alterados |
| `git add` | Move arquivos para a staging area |
| `git commit` | Salva um "ponto de restauração" no repositório local |
| `git log` | Exibe o histórico de commits |
| `git push` | Envia commits para o repositório remoto (GitHub) |
| `git pull` | Baixa atualizações do repositório remoto |
| `git clone` | Copia um repositório remoto para o computador |

> 🎮 **Analogia do videogame:** `git commit` é o "salvar o jogo". `git push` é sincronizar esse save com a nuvem. `git pull` é baixar o save de outro computador. `git clone` é baixar um jogo que ainda não está na sua máquina.

---

### Repositórios Remotos e o GitHub

O **GitHub** é uma plataforma na internet onde você armazena seus repositórios GIT. É o "Google Drive do código" — mas muito mais poderoso, pois guarda todo o histórico e permite colaboração.

> ☁️ **Analogia do Google Drive:** O repositório local é como um documento salvo só no seu computador. O repositório remoto no GitHub é como o mesmo documento no Google Drive — acessível de qualquer lugar, compartilhável com a equipe e com backup automático.

Para conectar seu repositório local ao GitHub, você cria um repositório vazio no site e aponta seu GIT local para esse endereço com o comando `git remote add origin`.

---

### Branches: trabalhando em paralelo

Uma **branch** (ramo) é uma linha de desenvolvimento paralela. Ela permite que você experimente mudanças sem afetar o código principal — e se as mudanças funcionarem, você as integra de volta.

> 🌱 **Analogia da planta:** A `main` é o tronco principal da árvore. Uma branch é um galho que cresce do tronco. Você pode podar o galho sem prejudicar a árvore. Se o galho crescer bem, você pode enxertá-lo de volta ao tronco (merge).

Por padrão, todo repositório começa com uma branch chamada `main` (antigamente chamada `master`). Você pode criar novos ramos com `git branch` e navegar entre eles com `git checkout`.

## 💻 Demonstração Prática — Código Comentado 

### Passo 1: Instalando o GIT

**Windows:**
```
Acesse https://git-scm.com/download/win e baixe o instalador.
Execute o instalador e clique em "Next" em todas as telas (padrão é suficiente).
Ao final, abra o "Git Bash" pelo menu Iniciar.
```

**macOS:**
```bash
# No terminal, execute:
xcode-select --install
# Ou instale via Homebrew:
brew install git
```

**Linux (Ubuntu/Debian):**
```bash
sudo apt update          # Atualiza a lista de pacotes
sudo apt install git -y  # Instala o GIT
```

**Verificando a instalação:**
```bash
git --version  # Exibe a versão instalada do GIT
```

```
git version 2.43.0
```

---

### Passo 2: Configuração inicial (uma vez por computador)

```bash
# Diz ao GIT quem você é — use seu nome e e-mail reais (os mesmos do GitHub)
git config --global user.name "Seu Nome Aqui"
git config --global user.email "seuemail@exemplo.com"

# Define o editor padrão (VS Code neste exemplo)
git config --global core.editor "code --wait"

# Define 'main' como nome padrão da branch principal (padrão moderno)
git config --global init.defaultBranch main

# Verifica todas as configurações salvas
git config --list
```

```
user.name=Seu Nome Aqui
user.email=seuemail@exemplo.com
core.editor=code --wait
init.defaultbranch=main
```

---

### Passo 3: Criando seu primeiro repositório local

```bash
# Cria uma pasta para o projeto (pode ser qualquer nome)
mkdir projeto-datajornada

# Entra na pasta criada
cd projeto-datajornada

# Inicializa o repositório GIT dentro da pasta
git init
```

```
Initialized empty Git repository in /home/usuario/projeto-datajornada/.git/
```

```bash
# Verifica o status do repositório (ainda vazio)
git status
```

```
On branch main

No commits yet

nothing to commit (create/copy files and use "git add" to track)
```

---

### Passo 4: Criando arquivos e fazendo o primeiro commit

```bash
# Cria um arquivo README com uma descrição do projeto
echo "# DataJornada — Projeto de Análise Exploratória" > README.md

# Verifica o status: o GIT detecta o novo arquivo
git status
```

```
On branch main

No commits yet

Untracked files:
  (use "git add <file>..." to include in what will be committed)
        README.md

nothing added to commit but untracked files present (use "git add" to track)
```

```bash
# Adiciona o arquivo à staging area (prepara para o commit)
git add README.md

# Verifica o status novamente: agora o arquivo está "staged"
git status
```

```
On branch main

No commits yet

Changes to be committed:
  (use "git rm --cached <file>..." to unstage)
        new file:   README.md
```

```bash
# Realiza o commit: salva um ponto de restauração com uma mensagem descritiva
git commit -m "Aula 01: criação do repositório e README inicial"
```

```
[main (root-commit) a3f8c2e] Aula 01: criação do repositório e README inicial
 1 file changed, 1 insertion(+)
 create mode 100644 README.md
```

```bash
# Exibe o histórico de commits
git log
```

```
commit a3f8c2e4b1d0f9e8c7b6a5d4e3f2c1b0a9d8e7f6 (HEAD -> main)
Author: Seu Nome Aqui <seuemail@exemplo.com>
Date:   Mon Apr 7 10:23:45 2025 -0300

    Aula 01: criação do repositório e README inicial
```

```bash
# Versão compacta do log (uma linha por commit — útil quando há muitos commits)
git log --oneline
```

```
a3f8c2e (HEAD -> main) Aula 01: criação do repositório e README inicial
```

---

### Passo 5: Adicionando mais arquivos e iterando

```bash
# Cria o notebook do projeto (arquivo vazio por enquanto)
touch analise_exploratoria.ipynb

# Adiciona TODOS os arquivos novos/modificados de uma vez
git add .                         # O ponto "." significa "tudo nesta pasta"

# Novo commit com mensagem descritiva
git commit -m "Aula 01: adiciona notebook inicial do projeto"

# Verifica o log com dois commits agora
git log --oneline
```

```
b7d1e4f (HEAD -> main) Aula 01: adiciona notebook inicial do projeto
a3f8c2e Aula 01: criação do repositório e README inicial
```

---

### Passo 6: Publicando no GitHub

```bash
# ANTES: crie um repositório VAZIO no GitHub
# Acesse github.com → "New repository" → dê um nome → NÃO marque "Initialize with README"
# Copie a URL do repositório criado (formato: https://github.com/seu-usuario/nome-repo.git)

# Conecta o repositório local ao repositório remoto no GitHub
git remote add origin https://github.com/seu-usuario/projeto-datajornada.git

# Verifica se o remote foi adicionado corretamente
git remote -v
```

```
origin  https://github.com/seu-usuario/projeto-datajornada.git (fetch)
origin  https://github.com/seu-usuario/projeto-datajornada.git (push)
```

```bash
# Envia os commits locais para o GitHub
# -u define o "upstream": na próxima vez, basta digitar "git push"
git push -u origin main
```

### Passo 7: Clonando um repositório existente

```bash
# Para baixar um repositório do GitHub para o seu computador
# (útil para pegar o projeto de um colega ou de um repositório público)
git clone https://github.com/algum-usuario/algum-projeto.git

# O GIT cria uma pasta com o nome do repositório e baixa tudo automaticamente
# Entre na pasta para trabalhar:
cd algum-projeto
```

### ⚠️ Erros Frequentes

**Erro 1: Esquecer de configurar o nome e e-mail antes do primeiro commit**

```bash
# ❌ ERRADO: fazer commit sem configurar o usuário
git commit -m "primeiro commit"
# Resultado: erro "Please tell me who you are"

# ✅ CORRETO: configurar antes de qualquer commit
git config --global user.name "Seu Nome"
git config --global user.email "email@exemplo.com"
git commit -m "primeiro commit"
```

---

**Erro 2: Fazer `git push` sem ter feito `git add` e `git commit` antes**

```bash
# ❌ ERRADO: tentar enviar ao GitHub sem commitar
# Você editou um arquivo e foi direto para:
git push origin main
# Resultado: "Everything up-to-date" — suas alterações NÃO foram enviadas!

# ✅ CORRETO: sempre seguir o fluxo completo
git add .
git commit -m "minha alteração"
git push origin main
```

---

**Erro 3: Usar `git add .` na pasta errada**

```bash
# ❌ ERRADO: executar git add na pasta raiz do computador por acidente
cd ~           # Foi para a pasta home sem querer
git add .      # Adicionaria TODOS os arquivos do computador!

# ✅ CORRETO: sempre verificar em qual pasta você está antes de qualquer comando
pwd            # Exibe a pasta atual — confira se é a pasta do projeto
cd projeto-datajornada
git add .
```

---

**Erro 4: Mensagem de commit vazia ou genérica demais**

```bash
# ❌ ERRADO: mensagens que não dizem nada
git commit -m "."
git commit -m "alterações"
git commit -m "atualizando"

# ✅ CORRETO: mensagens que descrevem O QUE foi feito e POR QUÊ
git commit -m "Aula 01: adiciona README com descrição do projeto DataJornada"
git commit -m "Aula 02: exploração numérica com NumPy na coluna de preços"
```

---

**Erro 5: Criar o repositório no GitHub com README e depois tentar dar push**

```bash
# ❌ PROBLEMA: se você marcou "Initialize with README" no GitHub,
# seu repositório remoto tem um commit que o local não tem.
# O git push vai falhar com "rejected - non-fast-forward"

# ✅ SOLUÇÃO: fazer pull primeiro para sincronizar os históricos
git pull origin main --allow-unrelated-histories
# Depois do merge, pode fazer push normalmente
git push origin main
```

### Exercício 1 — Fácil: Configuração e primeiro repositório

**Enunciado:** Configure o GIT com seu nome e e-mail. Em seguida, crie uma pasta chamada `meu-primeiro-repo`, inicialize um repositório GIT dentro dela, crie um arquivo `sobre-mim.txt` com seu nome e área de interesse em Data Science, adicione-o à staging area e faça o primeiro commit com uma mensagem descritiva.

> 💡 **Dica:** Use `echo "seu texto" > nome-do-arquivo.txt` para criar arquivos pelo terminal, ou crie pelo editor de texto normalmente.

<details>
<summary>Ver solução</summary>

```bash
# Passo 1: Configurar o GIT
git config --global user.name "Maria Silva"
git config --global user.email "maria@email.com"

# Passo 2: Criar a pasta e entrar nela
mkdir meu-primeiro-repo
cd meu-primeiro-repo

# Passo 3: Inicializar o repositório
git init

# Passo 4: Criar o arquivo
echo "Nome: Maria Silva | Interesse: Análise de dados de saúde pública" > sobre-mim.txt

# Passo 5: Adicionar à staging area
git add sobre-mim.txt

# Passo 6: Verificar o status (boa prática antes de commitar)
git status

# Passo 7: Fazer o commit
git commit -m "Aula 01: adiciona arquivo de apresentação pessoal"

# Passo 8: Confirmar no log
git log --oneline
```

```
(main) Aula 01: adiciona arquivo de apresentação pessoal
```

### Exercício 2 — Fácil: Iterando com commits

**Enunciado:** No repositório criado no exercício anterior, faça as seguintes alterações **em dois commits separados**:
1. Crie um arquivo `ideias.txt` com pelo menos 2 ideias de datasets para o projeto DataJornada.
2. Crie um arquivo `ferramentas.txt` listando as ferramentas que você pretende usar (Python, Pandas, etc.).

Ao final, exiba o histórico com `git log --oneline` e confirme que existem 3 commits no total.

> 💡 **Dica:** Cada arquivo deve ser um commit separado — não use `git add .` para adicionar os dois de uma vez.

<details>
<summary>Ver solução</summary>
